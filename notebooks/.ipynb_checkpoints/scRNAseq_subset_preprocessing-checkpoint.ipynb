```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib as mpl
```

```python
sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_versions()
sc.settings.set_figure_params(dpi=150, dpi_save= 300, fontsize=10, frameon=True, figsize=(2, 2), transparent=False, format="pdf")
```

```python
E125 = sc.read_10x_h5('./../../../data_10x/E12-5/outs/filtered_feature_bc_matrix.h5')
E145 = sc.read_10x_h5('./../../../data_10x/E14-5/outs/filtered_feature_bc_matrix.h5')
E165 = sc.read_10x_h5('./../../../data_10x/E16-5/outs/filtered_feature_bc_matrix.h5')
E185 = sc.read_10x_h5('./../../../data_10x/E18-5/outs/filtered_feature_bc_matrix.h5')
P15 = sc.read_10x_h5('./../../../data_10x/P1-5/outs/filtered_feature_bc_matrix.h5')
E125.var_names_make_unique()
E145.var_names_make_unique()
E165.var_names_make_unique()
E185.var_names_make_unique()
P15.var_names_make_unique()
```

```python
E125.obs["sample"] = "E-12.5"
E145.obs["sample"] = "E-14.5"
E165.obs["sample"] = "E-16.5"
E185.obs["sample"] = "E-18.5"
P15.obs["sample"] = "P-1.5"
```

```python
sc.external.pp.scrublet(E125)
sc.external.pp.scrublet(E145)
sc.external.pp.scrublet(E165)
sc.external.pp.scrublet(E185)
sc.external.pp.scrublet(P15)
```

```python
sc.external.pl.scrublet_score_distribution(E125)
sc.external.pl.scrublet_score_distribution(E145)
sc.external.pl.scrublet_score_distribution(E165)
sc.external.pl.scrublet_score_distribution(E185)
sc.external.pl.scrublet_score_distribution(P15)
```

```python
print(E125.shape)
print(E145.shape)
print(E165.shape)
print(E185.shape)
print(P15.shape)
```

```python
E125 = E125[E125.obs["doublet_score"] < 0.25,]
E145 = E145[E145.obs["doublet_score"] < 0.25,]
E165 = E165[E165.obs["doublet_score"] < 0.25,]
E185 = E185[E185.obs["doublet_score"] < 0.25,]
P15 = P15[P15.obs["doublet_score"] < 0.25,]
```

```python
print(E125.shape)
print(E145.shape)
print(E165.shape)
print(E185.shape)
print(P15.shape)
```

```python
samples = ["E-12.5", "E-14.5", "E-16.5", "E-18.5", "P-1.5"]
intestine_1 = E125.concatenate(E145, E165, E185, P15, batch_key= "sample", batch_categories=samples)
```

```python
intestine_1.obs_names_make_unique()
intestine_1.var_names_make_unique()
```

```python
sc.pp.filter_cells(intestine_1, min_genes=1)
sc.pp.filter_genes(intestine_1, min_cells=1)
intestine.shape
```

```python
intestine_1.var['mt'] = intestine_1.var_names.str.startswith('mt-')  # annotate the group of mitochondrial genes as 'mt'
sc.pp.calculate_qc_metrics(intestine_1, qc_vars=['mt'], percent_top=None, log1p=False, inplace=True)
```
